```python
import numpy
import scipy.signal
import toyplot

x = numpy.random.normal(size=(1000))

#alpha = 2 /(window + 1)
alpha = 0.05
b = [alpha]
a = [1, alpha-1]
zi = scipy.signal.lfiltic(b, a, x[0:1], [0])
xs = scipy.signal.lfilter(b, a, x, zi=zi)[0]

camvas, axes, mark = toyplot.plot(x, width=800, height=400)
axes.plot(xs)
```
